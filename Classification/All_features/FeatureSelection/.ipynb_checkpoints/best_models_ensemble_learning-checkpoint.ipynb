```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
import numpy as np

# Load the test results from the CSV file
test_results_df = pd.read_csv("test_results_PR.csv")
TITLE='PR'
# List of feature selection techniques
feature_selection_techniques = [
    "SelectKBest_f_regression",
    "SelectKBest_mutual_info_regression",
    "VarianceThreshold"
]

# Dictionary to store the best results for each feature selection technique
best_results = {}

# Set to store all unique features
all_features = set()

# Iterate over each feature selection technique
for technique in feature_selection_techniques:
    # Filter results for the current technique
    technique_results = test_results_df[test_results_df['Selection Method'] == technique]
    
    # Select the top three results based on lowest RMSE
    top_three_rows = technique_results.nsmallest(3, 'RMSE')
    
    # Iterate over the top three rows
    for idx, best_row in top_three_rows.iterrows():
        # Extract actual and predicted values
        actual_values = ast.literal_eval(best_row['Actual'])
        predicted_values = ast.literal_eval(best_row['Predicted'])
        features = ast.literal_eval(best_row['Selected Features'])

        # Add the features to the set of all features
        all_features.update(features)
        
        # Store the best result
        if technique not in best_results:
            best_results[technique] = []
        best_results[technique].append({
            'RMSE': best_row['RMSE'],
            'Actual': actual_values,
            'Predicted': predicted_values,
            'Features': features
        })

# Colors and markers for different techniques
colors = ['blue', 'green', 'orange']
markers = ['o', 's', '^']

# Plot the actual vs. predicted values for the best cases
plt.figure(figsize=(12, 8))

for idx, (technique, results_list) in enumerate(best_results.items()):
    for i, results in enumerate(results_list):
        plt.scatter(results['Actual'], results['Predicted'], label=f"{technique} (RMSE: {results['RMSE']:.4f})", color=colors[idx], marker=markers[i],linewidths=0.15)

# Calculate ensemble predictions
ensemble_actual = best_results[feature_selection_techniques[0]][0]['Actual']
ensemble_predicted = np.mean(
    [results['Predicted'] for results_list in best_results.values() for results in results_list],
    axis=0
)

# Add the ensemble approach to the plot
plt.scatter(ensemble_actual, ensemble_predicted, label="Ensemble (1/9 weight each)", color='red', marker='x',linewidths=1)

# Plot a reference line
min_val = min(min(ensemble_actual), min(ensemble_predicted))
max_val = max(max(ensemble_actual), max(ensemble_predicted))
plt.plot([min_val, max_val], [min_val, max_val], 'k--', linewidth=2)

# Add labels and legend
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values for Best Feature Selection Techniques and Ensemble '+TITLE)
plt.legend()
plt.grid(True)
plt.tight_layout()

# Save plot
plt.savefig("actual_vs_predicted_best_feature_selection_and_ensemble_"+TITLE+".png")

# Show plot
plt.show()
```
